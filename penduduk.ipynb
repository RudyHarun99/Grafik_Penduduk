```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
```

```python
df = pd.read_excel('penduduk.xlsx',header=3,skipfooter=3,na_values='-')
df = df.fillna(0)
df = df.set_index('Unnamed: 0')
df
```

```python
fig=plt.figure(figsize=(16,8))
p=plt.subplot(111,projection='3d')
x=np.arange(len(df.index))
y=np.arange(len(df.index))
z=np.zeros(len(df.index))
dx=np.ones(len(df.index))
dy=np.ones(len(df.index))
dz=df[1971]
p.bar3d(x,y,z,dx,dy,dz)
p.set_xticks(np.arange(len(df.index)))
p.set_xticklabels(df.index,rotation=90)
p.set_zlabel('Populasi')
plt.show
```
